### Computing mean and std for each station/year ###
And calculating the standard deviation. In this case we are not divi

```python
# a simple scale function to normalize the data-frame row-by-row
from numpy import mean, std
def scale_temps(Din):
    matrix=Din.iloc[:,3:]
    Dout=Din.loc[:,['station','year']+range(1,366)]
    Mean=mean(matrix, axis=1).values
    Dout['Mean']=Mean
    Std= std(matrix, axis=1).values
    Dout['Std']=Std
    # Decided not to normalize each year to have mean zero and std 1
    # tmp = pd.DataFrame((matrix.values - Mean[:,np.newaxis])/Std[:,newaxis],columns=range(1,366))
    # print tmp.head()
    Dout.loc[:,1:365]=matrix.values
    return Dout
Dout=scale_temps(Data)
#reorder the columns
Dout=Dout[['station','year','Mean','Std']+range(1,366)]
Dout.head()
```

#### SuperGeomDensity2D : Custom class for making rplot with more contour lines

```python
import matplotlib.pyplot as plt
from pandas.tools import rplot

class SuperGeomDensity2D(rplot.Layer):
    def __init__(self,data=None,**kwds):
        rplot.Layer.__init__(self,data,**kwds)
        self.ccount=kwds.get('ccount',15) #ccount = count of contour lines
        
    def work(self, fig=None, ax=None):
        """Draw a two dimensional kernel density plot.
        You can specify either a figure or an axis to draw on.

        Parameters:
        -----------
        fig: matplotlib figure object
        ax: matplotlib axis object to draw on

        Returns:
        --------
        fig, ax: matplotlib figure and axis objects
        """
        if ax is None:
            if fig is None:
                return fig, ax
            else:
                ax = fig.gca()
        x = self.data[self.aes['x']]
        y = self.data[self.aes['y']]
        rvs = np.array([x, y])
        x_min = x.min()
        x_max = x.max()
        y_min = y.min()
        y_max = y.max()
        X, Y = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
        positions = np.vstack([X.ravel(), Y.ravel()])
        values = np.vstack([x, y])
        import scipy.stats as stats
        kernel = stats.gaussian_kde(values)
        Z = np.reshape(kernel(positions).T, X.shape)
        ax.contour(X,Y,Z, self.ccount)
        return fig, ax
```

```python
#Plotting a figure with 15 contour lines
X='latitude';Y='Mean'
dfTmp=Djoined[[X,Y]].dropna()
dfTmp=dfTmp.iloc[:,:]
p = rplot.RPlot(dfTmp,x=X,y=Y)
#p.add(rplot.GeomPoint()) Using this causes IndexError: tuple index out of range
p.add(rplot.GeomScatter())
p.add(SuperGeomDensity2D())
p.render();
```

```python
def plot_reconstructions(selection,rows=2,columns=7):
    Recon=Eig*Prod.transpose()+Mean[:,np.newaxis]
    plt.figure(figsize=(columns*3,rows*3),dpi=300)
    j=0;
    for i in selection:
        subplot(rows,columns,j); 
        j += 1; 
        if j>=rows*columns: break
        plot(Recon[:,i])
        plot(Djoined.ix[i,1:365]);
        title(Djoined.ix[i,'station']+' / '+str(Djoined.ix[i,'year']))
        xlim([0,365])

from __future__ import division
def recon_error(selection, Eig, Prod):
    Recon = Eig * Prod.transpose() + Mean[:,np.newaxis]
    j = 0
    error = np.zeros(shape(selection))
    for i in selection:
        for k in range(365):
            error[j] += np.square(Djoined.ix[i,k+1]-Recon[k,i])
        error[j] = error[j] / 365
        j+=1
    return np.mean(error[isnan(error)==False])
```

<span style="color:red"> Check how the approximations change/improve as you increase the number of coefficients</span>

We find that the reconstruction error reduces as the number of coefficients is increased

```python
def err(V):
    """ err = the average of the square values """
    return np.dot(V,V.T)/len(V)

def meanErr(D):
    """ meanErr = the average err over all rows of D """
    total_err=0
    n=shape(D)[0]
    for i in range(n):
        total_err += err(D[i,:])
    return total_err/n

def recon_error(U, k, D):
    '''
        Outputs the reconstruction error using the top k eigen vectors and returns the corresponding reconstruction
        U = all eigenvectors
        k = number of eigenvectors to use
        D = original data
    '''
    Eig=np.matrix(U[:,:k])
    #print [np.linalg.norm(U[:,i]) for i in range(k)] #norm of eigenvectors
    m=np.matrix(D)
    m[isnan(m)]=0
    matrix = m-Mean
    Prod=matrix*Eig;
    Recon=Eig*Prod.transpose()+Mean[:,np.newaxis]
    Error=meanErr(Recon.transpose()-m)
    print "Reconstruction error is: " + str(Error)
    return Prod, Recon, Error
```
