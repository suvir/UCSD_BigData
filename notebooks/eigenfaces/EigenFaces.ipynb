### Compute the average lengths of vectors ###

```python
def err(V):
    """ err = the average of the square values """
    return np.dot(V,V)/len(V)

def meanErr(D):
    """ meanErr = the average err over all rows of D """
    total_err=0
    n=shape(D)[0]
    for i in range(n):
        total_err += err(D[i,:])
    return total_err/n
print 'before subtracting mean mean err=',meanErr(D0)
print 'after subtracting mean mean err=',meanErr(D1)
```

### Using iterative PCA to find the largest eigen-vector ###

```python
from numpy.linalg import norm

k=60; L=10

R=copy(D1)   #R is a matrix in which each row is the residual vector after the current eigenvectors have been projected out.
eigvecs=np.zeros([k,H*W])
Coeffs=np.zeros([im_number,k])  # Coeffs holds the coefficients of the eigenvectors that approximate each image
var_remaining=np.zeros(k)

factor=im_number*H*W

for e in range(k):
    print 'starting on eigenvec',e
    V=np.random.normal(size=H*W)  # V is the "current vector" in the eigenvector finding algorithm.
    V=V/norm(V)

    # Iterate L times through the data to find the next eigenvector.
    for j in range(L):
        acc_length=0; acc_explained=0
        Vprev=V/norm(V)
        for i in range(im_number):
            X=R[i,:]
            d=np.dot(V,X)/norm(V)
            V=V+d*X
            #print acc_length,acc_explained
            acc_length += np.dot(X,X)
            acc_explained += d*d
        NV=V/norm(V) # normalized version of updated vector
        print '%4d: change=%5.4f, total variance=%5.4f, captured variance=%5.4f' % (j,norm(NV-Vprev),acc_length/factor,acc_explained/factor)
    var_remaining[e]=acc_length/factor
    eigvecs[e,:]=NV
    # subtract projection on V from each data vector
    for i in range(im_number):
        X=R[i,:]
        d=np.dot(X,NV)
        Coeffs[i,e]=d
        X=X-NV*d
        R[i,:]=X
```

For each new eigenvector $\mathbf{Eigen}$, we can calculate it as follow,
 
$\mathbf{Eigen}$ = a random vector of length $p$

Do Max Iteration times:

1 $\mathbf{sum}$ = 0 (a vector of length p)
      
2 for each row $\mathbf{x} \in \mathbf{X}$ : $\mathbf{sum} = \mathbf{sum} + (\mathbf{x} \cdot \mathbf{Eigen})\mathbf{x}$
            
3 $\mathbf{Eigen} = \frac{\mathbf{sum}}{|\mathbf{sum}|}$
      
return $\mathbf{Eigen}$

The main reason is following, define $$cov = \frac{1}{n} \sum x x^T$$

For a random vector r, $sum = cov \cdot r^k$, where k is iteration time. Therefore, we must initialize the $sum = 0$ and let $Eigen = sum = cov \cdot r^k$ when the new iteration begins, and we can get the $sum = cov \cdot r^(k+1).$

```python
# the function implement the iterative pca algorithm
def iterative_pca(X, eigen_num, max_iteration = 10):  
    [example_num, dimen] = X.shape
    eigvecs = np.zeros([eigen_num, dimen])
    coeffs = np.zeros([example_num, eigen_num]) 
    print 'Calculating eigenvectors .............'
    for e in range(eigen_num):
        print e, '........',
        next_eigen = np.random.normal(size=dimen)
        next_eigen = next_eigen / norm(next_eigen)
        # Iterate L times through the data to find the next eigenvector.
        for j in range(max_iteration):
            sum_vector = np.zeros(dimen)
            for i in range(im_number):
                xi = X[i,:]
                sum_vector += np.dot(next_eigen, xi) * xi
            next_eigen = sum_vector / norm(sum_vector)
            
        eigvecs[e,:] = next_eigen
        # subtract projection on V from each data vector
        for i in range(example_num):
            xi = X[i,:]
            d = np.dot(xi, next_eigen)
            coeffs[i, e] = d
            xi = xi - next_eigen*d
            X[i,:] = xi
    return [eigvecs, coeffs]
```
